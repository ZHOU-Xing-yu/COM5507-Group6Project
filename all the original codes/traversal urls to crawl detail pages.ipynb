```python
import requests
import openpyxl
import pandas as pd
import bs4 as bs
from lxml import etree
#定义一个空列表
stu_num=[]
workbook1=openpyxl.load_workbook(r'/Users/zhouxingyu/Desktop/PYTHONSTUDY/joblinksnew.xlsx')
#选定目标sheet
worksheet1 = workbook1.active
for cell in worksheet1['B']:
    #print(cell.value)
    stu_num.append(cell.value)
    # for item in stu_num:
    #      if "公司介绍页链接" in item:
    #          stu_num.remove(item)
print(stu_num)
```

```python
detail=[]
for i in stu_num:
    response = requests.get(i)
    detail.append(response.content)
print(detail)
```

```python
details=[]
for i in detail:
    page_html = bs.BeautifulSoup(i, 'html.parser')
    DETAILS = page_html.find_all('div', class_='text-dark-grey-text p-2 p-md-0')
    for i in DETAILS:
        details.append(i.get_text(strip=True))
details
# str6 = " ".join(detail)
# str6
```

```python
data2022 = {
    'DetailInfo': details
}
```

```python
import pandas as pd 
pd_webcareer = pd.DataFrame.from_dict(data2022)
```

```python
pd_webcareer.to_excel('DETAIL.xlsx',header=None,index=False)
```
